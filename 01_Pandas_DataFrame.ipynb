# Bekerja Dengan Data yang Hilang

```python
import numpy as np
```

```python
nilai = {
    'Nilai pertama': [100, 90, np.nan, 95],
    'Nilai kedua': [30, 45, 56, np.nan],
    'Nilai ketiga': [np.nan, 80, 90, np.nan]
}
```

```python
df_nilai = pd.DataFrame(nilai)
df_nilai
```

```python
# Mengecek data yang hilang
df_nilai.isnull()
```

```python
# Mengecek data yang tidak hilang
df_nilai.notnull()
```

```python
# Mengisi nilai yang hilang / kosong dengan fungsi fillna()
df_nilai.fillna(0)
```

```python
nilai2 = {
    'Nilai pertama': [100, 90, 78, 95],
    'Nilai kedua': [30, 45, 56, 83],
    'Nilai ketiga': [np.nan, 80, 90, np.nan],
    'Nilai keempat': [87, 56, 49, 90]
}
```

```python
df_nilai2 = pd.DataFrame(nilai2)
df_nilai2
```

```python
# Menghapus nilai yang hilang
df_nilai2.dropna()
```
